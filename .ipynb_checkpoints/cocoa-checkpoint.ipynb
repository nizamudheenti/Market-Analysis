```python
import pandas as pd
```

```python
data=pd.read_csv('pinksheet.csv')
```

```python
data1=data[1:721]
data2=data1[['Unnamed: 0','Cocoa']]
data2=data2.rename(columns={"Unnamed: 0":"date"})
data2.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=[20,10])
sns.scatterplot(data=data2[650:],x='date',y='Cocoa')
plt.xlabel("MONTH")
plt.ylabel('PRICE)')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(data2['date'])
data2['date']=le.transform(data2['date'])
```

```python
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(np.asmatrix(data2['date']).reshape(-1,1),np.asmatrix(data2['Cocoa']).reshape(-1,1))
```

```python
data3=[731]
pre=lr.predict(np.asmatrix(data2['date']).reshape(-1,1))
from sklearn.metrics import mean_squared_error
```

```python
#pre
mean_squared_error(np.asmatrix(data2['Cocoa']).reshape(-1,1),pre)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor(n_estimators=250,max_depth=8,random_state=1111)
```

```python
X=np.asmatrix(data2['date']).reshape(-1,1)
y=np.asmatrix(data2['Cocoa']).reshape(-1,1)
```

```python
rfr.fit(X,y)
```

```python
pre2=rfr.predict(np.asmatrix(data2['date']).reshape(-1,1))
mean_squared_error(np.asmatrix(data2['Cocoa']).reshape(-1,1),pre2)
```

```python
pred3=rfr.predict(X)
#pd.DataFrame(y,pred3)
plt.plot(X, y, 'b-',
         label='Actual')
plt.plot(X, pred3, 'r-',
         label='predicted')
plt.legend(loc='upper right')
plt.xlabel('MONTH')
plt.ylabel('PRIZE')

plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gr=GradientBoostingRegressor(loss='quantile', alpha=0.6,
                                n_estimators=250, max_depth=8,
                                learning_rate=0.15)
```

```python
gr.fit(X,y)
```

```python
pred4=gr.predict(X)
```

```python
mean_squared_error(y,pred4)
```

```python
plt.plot(X, y, 'b-',
         label='Actual')
plt.plot(X, pred4, 'r-',
         label='predicted')
plt.legend(loc='upper right')
plt.xlabel('MONTH')
plt.ylabel('PRIZE')

plt.show()
```

```python
from sklearn.svm import SVR
```

```python
svr=SVR(C=1.0, epsilon=0.2)
```

```python
svr.fit(X,y)
```

```python
pred5=svr.predict(X)
```

```python
plt.plot(X, y, 'b-',
         label='Actual')
plt.plot(X, pred5, 'r-',
         label='predicted')
plt.legend(loc='upper right')
plt.xlabel('MONTH')
plt.ylabel('PRIZE')

plt.show()
```

```python
svr_rbf = SVR(kernel='rbf', C=1, gamma=0.1, epsilon=.1)
svr_lin = SVR(kernel='linear', C=1, gamma='auto')
svr_poly = SVR(kernel='poly', C=1, gamma='auto', degree=3, epsilon=.1,
               coef0=1)
```

```python
svr_rbf.fit(X,y)
svr_lin.fit(X,y)
svr_poly.fit(X,y)
```

```python
rbf=svr_rbf.predict(X)
lin=svr_lin.predict(X)
poly=svr_poly.predict(X)
```

```python
from sklearn.linear_model import BayesianRidge,ARDRegression
clf = BayesianRidge(compute_score=True)
clf.fit(X, y)
pred6=clf.predict(X)
```

```python
plt.plot(X, y, 'b-',
         label='Actual')
plt.plot(X, pred6, 'r-',
         label='predicted')
plt.legend(loc='upper right')
plt.xlabel('MONTH')
plt.ylabel('PRIZE')

plt.show()
```

```python
clf = ARDRegression()
clf.fit(X, y)
```

```python
pred7=clf.predict(X)
plt.plot(X, y, 'b-',
         label='Actual')
plt.plot(X, pred7, 'r-',
         label='predicted')
plt.legend(loc='upper right')
plt.xlabel('MONTH')
plt.ylabel('PRIZE')

plt.show()
```
